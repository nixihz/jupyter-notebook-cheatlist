```python
import pandas as pd
import matplotlib.pyplot as plt


# 解决 pandas 中文不显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']


```

```python
# 读取数据
df = pd.read_csv("input.csv")

# jupyter notebook 每个节点最后一行表示输出
df
```

```python
# df 输出图表，keyword 作为横坐标，count 作为纵坐标
df.plot(x='keyword', y='count', kind='bar', figsize=(15, 7), fontsize=14, rot=75)
```

```python
# count > 100000 的数据
df2 = df[df["count"] > 100000]

df2
```

```python
df2.to_csv("output.csv", index=False)
```
